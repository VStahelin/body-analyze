```python
data_frame['7day_rolling_avg'] = data_frame.calories.rolling(7).mean()
display(data_frame.head(10)) 
```

```python
after7_data_frame = data_frame.query('index > 27')
after7_data_frame['7day_rolling_avg'] = after7_data_frame.calories.rolling(7).mean()
display(after7_data_frame)

plt.figure(figsize=(30,5))
sns.lineplot(
    data=after7_data_frame,
    label="Calories rate",
    x="date",
    y="calories"
)
sns.lineplot(
    data=after7_data_frame,
    label="Calories rolling avg",
    x="date",
    y="7day_rolling_avg"
)

plt.xlabel( 'Months')
plt.ylabel('Calories') 
pos = ['2022-07-31', '2022-08-30', '2022-09-30', '2022-10-30','2022-11-30'] 
lab = ['July', 'Aug', 'Sept', 'Oct', 'Nov'] 
plt.xticks( pos, lab)
 
```
