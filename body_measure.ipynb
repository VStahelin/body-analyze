```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

data_frame = pd.read_csv('data/BODY.csv')
data_frame = data_frame.dropna().reset_index(drop=True)
data_frame.time = (pd.to_datetime(data_frame.time)).dt.strftime('%m-%d')
data_frame.drop_duplicates(subset=['time'], keep='last', inplace=True)


plt.figure(figsize=(12,5))
sns.lineplot(
    data=data_frame,
    label="muscleRate",
    x="time",
    y="muscleRate"
)
sns.lineplot(
    data=data_frame,
    label="bmi",
    x="time",
    y="bmi"
)
sns.lineplot(
    data=data_frame,
    label="bodyWaterRate",
    x="time",
    y="bodyWaterRate"
)
sns.lineplot(
    data=data_frame,
    label="visceralFat",
    x="time",
    y="visceralFat"
)
# sns.lineplot(
#     data=data_frame,
#     label="metabolism",
#     x="time",
#     y="metabolism"
# )

```
